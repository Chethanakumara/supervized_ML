```python
import requests   # for making API calls
import pandas as pd  # for working with data
```

```python
url = "https://gbfs.citibikenyc.com/gbfs/en/station_information.json"
```

```python
response = requests.get(url)
```

```python
# Convert JSON response to Python dictionary
data = response.json()
```

```python
data
```

```python
# Suppose the JSON data is stored in variable 'data'
stations = data['data']['stations']
```

```python
# Convert the list of station dictionaries to a DataFrame
df = pd.DataFrame(stations)

```

```python
df
```

```python
selected_col= df[['name', 'lat', 'lon', 'capacity']]
```

```python
dd=selected_col.iloc[:1500]
```

```python
dd
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
```

```python
x = dd[['lat','lon']]
y = dd[['capacity']]
```

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y , test_size=0.2 ,random_state=43)
```

```python
m=LinearRegression()
m.fit(x_train,y_train)
```

```python
pred = m.predict(x_test)
pred
```

```python
from flask import Flask, request, jsonify
import joblib
import pandas as pd

app = Flask(__name__)

# Load the trained model
model = joblib.load("saveit_as.pkl")

@app.route("/")
def home():
    return " Bike Station Capacity Prediction API is Running!"

# Endpoint: /predict
@app.route("/predict", methods=["POST"])
def predict_capacity():
    try:
        # Get JSON data from request body
        data = request.get_json()
        lat = data.get("lat")
        lon = data.get("lon")

        # Validate input
        if lat is None or lon is None:
            return jsonify({"error": "Please provide both lat and lon"}), 400

        # Create DataFrame for model
        input_data = pd.DataFrame({"lat": [lat], "lon": [lon]})

        # Predict
        prediction = model.predict(input_data)[0]

        # Return JSON response
        return jsonify({
            "latitude": lat,
            "longitude": lon,
            "predicted_capacity": round(float(prediction), 2)
        })

    except Exception as e:
        return jsonify({"error": str(e)}), 500


if __name__ == "__main__":
    app.run(debug=True)
```
